```python
#interquartile - remove lesser and greater outlier 
```

```python
dash=pd.DataFrame(index=['Mean',"Median","Mode","Q1:25%","Q2:50%","Q3:75%",'99%',"Q4:100%",'IQR','1.5rule','Lesser','Greater','Min',"Max"],columns=quan)
for colum in quan:
    dash[colum]['Mean']=dataset[colum].mean()
    dash[colum]['Median']=dataset[colum].median()
    dash[colum]['Mode']=dataset[colum].mode()[0]
    dash[colum]['Q1:25%']=np.percentile (dataset[colum],25)
    dash[colum]['Q2:50%']=np.percentile (dataset[colum],50)
    dash[colum]['Q3:75%']=np.percentile (dataset[colum],75)
    dash[colum]['99%']=np.percentile (dataset[colum],99)
    dash[colum]['Q4:100%']=np.percentile (dataset[colum],100)
    dash[colum]['IQR']=dash[colum]['Q3:75%']-dash[colum]['Q1:25%']
    dash[colum]['1.5rule']=1.5*dash[colum]['IQR']
    dash[colum]['Lesser']=dash[colum]['Q1:25%']-dash[colum]['1.5rule']
    dash[colum]['Greater']=dash[colum]['Q3:75%']+dash[colum]['1.5rule']
    dash[colum]['Min']=dash[colum].min()
    dash[colum]['Max']=dash[colum].max()
```
