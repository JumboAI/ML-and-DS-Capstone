```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot with regression line
sns.regplot(x='Item_Visibility', y='Item_MRP', data=dataset, scatter_kws={'alpha': 0.5})
plt.title('Scatter Plot: Item_Visibility vs. Item_MRP')
plt.xlabel('Item_Visibility')
plt.ylabel('Item_MRP')
plt.show()

# Histograms for individual distributions
dataset['Item_Visibility'].hist(bins=20, alpha=0.7, label='Item_Visibility')
plt.legend()
plt.title('Distribution of Item_Visibility')
plt.xlabel('Item_Visibility')
plt.ylabel('Frequency')
plt.show()

dataset['Item_MRP'].hist(bins=20, alpha=0.7, label='Item_MRP')
plt.legend()
plt.title('Distribution of Item_MRP')
plt.xlabel('Item_MRP')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.pairplot(dataset)
```

```python
from scipy.stats import pearsonr

# Assuming with two coloum
correlation, p_value = pearsonr(dataset['Item_Weight'], dataset['Item_MRP'])

print(f"Pearson correlation coefficient: {correlation}")
print(f"P-value: {p_value}")
```

```python
cov_matrix = dataset.cov()
print (cov_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualize the covariance matrix using a heatmap
sns.heatmap(cov_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Covariance Matrix Heatmap')
plt.show()
```
