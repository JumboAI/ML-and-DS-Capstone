```python
#create histogram for both before and after for all the coloum
import seaborn as sns
import matplotlib.pyplot as plt

def plot_histograms(dataset, columns, title_suffix):
    for column in columns:
        plt.figure(figsize=(8, 5))
        sns.histplot(dataset[column], kde=True, color='green', line_kws={'linewidth': 2})
        plt.title(f'Histogram of {column} {title_suffix}')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.show()

# Columns to analyze
quan = ['Item_Weight', 'Item_Visibility', 'Item_MRP', 'Item_Outlet_Sales']

# Before outlier handling
plot_histograms(dataset, quan, '(Original Data)')

# After outlier handling
plot_histograms(dash, quan, '(Cleaned Data)')
```

```python
#two sets of boxplots: one for the original data and one for the cleaned data 
#detail explanation given in pdf file

import seaborn as sns
import matplotlib.pyplot as plt

def plot_boxplots(dataset, columns, title_suffix):
    for column in columns:
        plt.figure(figsize=(8, 5))
        sns.boxplot(data=dataset, x=column, color='lightgreen')
        plt.title(f'Boxplot of {column} {title_suffix}')
        plt.xlabel(column)
        plt.ylabel('Values')
        plt.show()

# Plot boxplots for the columns
plot_boxplots(dataset, quan, '(Original Data)')
plot_boxplots(dash, quan, '(Cleaned Data)')
```

```python
def get_pdf_probability(dataset,startrange,endrange):
    from matplotlib import pyplot
    from scipy.stats import norm
    import seaborn as sns
    ax=sns.distplot(dash,kde=True,kde_kws={'color':'blue'},color ='blue')
    pyplot.axvline(startrange,color='Red')
    pyplot.axvline(endrange,color='Red')
    #generate a sample
    sample=dataset
    #calculate parameters
    sample_mean=sample.mean()
    sample_std=sample.std()
    print('Mean=%.3f, Standard Deviation=%3.f'%(sample_mean,sample_std))
    #define the distribution
    dist=norm(sample_mean,sample_std)
    #sample probabilities for a range of outcomes
    values=[value for value in range(startrange,endrange)]
    probabilities=[dist.pdf(value) for value in values]
    prob=sum(probabilities)
    print("The area between range({},{}):{}".format(startrange,endrange,sum(probabilities)))
    return prob
```

```python
def stdNBgraph(dataset):
    #Covert to Standard Normal Distribution
    import seaborn as sns
    mean=dataset.mean()
    std=dataset.std()
    
    values=[i for i in dataset]
    z_score=[((j-mean)/std) for j in values]
    sns.distplot(z_score,kde=True)
    sum(z_score)/len(z_score)
    #z_score.std()
```

```python
stdNBgraph(dash["Item_Outlet_Sales"])
```
